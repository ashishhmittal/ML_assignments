```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
df.head()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df=df.fillna(df.mean())
df.isnull().sum()
df=df.dropna()
df.head()
```

```python
df.shape
```

```python
df.columns
```

# 1. Plot Between Pressure  and Temperature at 9 am

```python
with sns.axes_style('white'):
    sns.jointplot(x='Temp9am', y='Pressure9am',kind='hex' ,data=df,color='k')
```

# 2. Plot Between Pressure and Temperature at 3 pm

```python
with sns.axes_style('white'):
    sns.jointplot(x='Temp3pm', y='Pressure3pm',kind='hex' ,data=df,color='k')
```
